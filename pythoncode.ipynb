```python
#import pandas library and provide alias name as pd
import pandas as pd
#Data from: https://www.kaggle.com/shivamb/netflix-shows
netflix_data = pd.read_csv('Desktop/netflix_titles.csv')
netflix_data
```

```python
n_data = netflix_data.groupby(['release_year', 'type'], as_index=False).show_id.count()
n_data.columns = ['release_year', 'type', 'count']
n_data.head()
```

```python
n_data_pivot = n_data.pivot(index='release_year', columns='type', values='count').reset_index()
n_data_pivot.fillna(0, inplace = True)
n_data_pivot.head()
```

```python
# Declare a chart with one Axis
fig, ax = plt.subplots()
# Creating some dummy data
ax.plot(n_data_pivot.release_year, n_data_pivot.Movie, label = 'Movies')
ax.plot(n_data_pivot.release_year, n_data_pivot['TV Show'], label = 'Tv Shows')
ax.set_ylabel('Number of Movies/TV Shows')
ax.set_xlabel('Year of Release')
ax.set_title("Trend Chart")
plt.text(x=1970, y=400, s=r'Majority Movies are released after 2000 and Tv Shows mostly after 2010')#, fontsize = 12)
fig.set_size_inches(18.5, 10.5)
plt.grid()
ax.legend()
```

```python
fig1, ax1 = plt.subplots()
#plt.scatter(n_data_pivot['Movie'], n_data_pivot['TV Show'])
plt.scatter(n_data_pivot['Movie'], n_data_pivot['TV Show'], color='red')
ax1.set_xlabel('Number of Movies in a given Year')
ax1.set_ylabel('Number of TV Shows in a given Year')
ax1.set_title("Relationship between Number of Movies and TV Shows")
fig1.set_size_inches(15, 10)
plt.rcParams.update({'font.size': 10})
plt.grid()
plt.show()
```

```python
# Declare a chart with one Axis
fig, ax = plt.subplots()
# Creating some dummy data
ax.plot(n_data_pivot.release_year, n_data_pivot.Movie, label = 'Movies')
ax2 = ax.twinx()
ax2.plot(n_data_pivot.release_year, n_data_pivot['TV Show'], label = 'Tv Shows', color = 'red')
ax.set_ylabel('Number of Movies')
ax2.set_ylabel('Number of TV Shows')
ax.set_xlabel('Year of Release')
ax.set_title("Trend Chart")
fig.set_size_inches(18.5, 10.5)
plt.grid()
ax.legend(loc="upper left")
ax2.legend(loc="upper right")
```
